# **Assignment 01**

```python


import numpy as np

# Creating two random 3x3 matrices
matrix1 = np.random.randint(0, 10, size=(3, 3))
matrix2 = np.random.randint(0, 10, size=(3, 3))

# Matrix Addition
matrix_sum = matrix1 + matrix2

# Matrix Subtraction
matrix_diff = matrix1 - matrix2

# Matrix Multiplication
matrix_product = np.dot(matrix1, matrix2)

# Determinants
det1 = np.linalg.det(matrix1)
det2 = np.linalg.det(matrix2)

# Inverse of matrix1 (if possible)
if det1 != 0:
    inverse1 = np.linalg.inv(matrix1)
else:
    print("Matrix 1 is singular, inverse not possible")

# Inverse of matrix2 (if possible)
if det2 != 0:
    inverse2 = np.linalg.inv(matrix2)
else:
    print("Matrix 2 is singular, inverse not possible")

# Rank calculation
rank1 = np.linalg.matrix_rank(matrix1)
rank2 = np.linalg.matrix_rank(matrix2)

# Display results
print("Matrix 1:\n", matrix1, "\n")
print("Matrix 2:\n", matrix2, "\n")
print("Addition Result:\n", matrix_sum, "\n")
print("Subtraction Result:\n", matrix_diff, "\n")
print("Multiplication Result:\n", matrix_product, "\n")
print("Determinant of Matrix 1:", det1, "\n")
print("Determinant of Matrix 2:", det2, "\n")

if det1 != 0:
    print("Inverse of Matrix 1:\n", inverse1, "\n")

if det2 != 0:
    print("Inverse of Matrix 2:\n", inverse2, "\n")

print("Rank of Matrix 1:", rank1)
print("Rank of Matrix 2:", rank2)
```

# **Assignment 02**

```python
# Random Matrix with float
A = np.random.uniform(11,30, size=(5,3))
B = np.random.uniform(31,40, size=(5,3))

print("Matrix A: \n",A,"\n")
print("Matrix B: \n",B,"\n")

# Flattem 1D Array
A_flat = A.flatten()
B_flat = B.flatten()

print("Flattened Matrix A: \n",A_flat,"\n")
print("Flattened Matrix B: \n",B_flat,"\n")

# Create an Index for x-Axis
x = np.arange(len(A_flat))

# Line Plot
plt.plot(x, A_flat, marker = "o", linestyle = "-", label="Matrix A")
plt.plot(x, B_flat, marker = "s", linestyle = "--", label="Matrix B")

# Label and Title
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Line Graph of Float Flatten Matrices A and B")
plt.legend()
plt.grid(True)

# Show the Plot
plt.show()
```
